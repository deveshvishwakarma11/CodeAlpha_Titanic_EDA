```python
import matplotlib.pyplot as plt
```

```python
df["Survived"].value_counts().plot(kind="bar")       # Survival count graph.

plt.title("Survival Count")
plt.xlabel("Survived (0=No, 1=Yes)")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
df.groupby("Sex")["Survived"].mean().plot(kind="bar")       # Male vs Female survival rate.

plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
df.groupby("Pclass")["Survived"].mean().plot(kind="bar")       # Passenger survival rate by class

plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
df["Age"].hist(bins=20)       # Age distribution

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
df["Fare"].hist(bins=20)       # Fare distrubution

plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")
plt.show()
```

# Titanic Dataset - Key Insights

1. Dataset contains 891 passengers and 12 features.

2. Age, Cabin and Embarked columns contain missing values.

3. Overall survival rate was approximately 38%.

4. Female passengers had a much higher survival rate than male passengers.

5. First-class passengers had a higher chance of survival than second and third-class passengers.

6. Most passengers were between 20–40 years old.

7. Most passengers paid lower fares, while a few paid very high fares.
